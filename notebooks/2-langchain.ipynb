## Setting temperature, max tokens and streaming

* `max_completion_tokens`: Limits the extension of the response

* `temperature`: Response randomness: $\in [0,2]$

* `stream`: Prints out the output continuously

````python
# Creating a sarcastic chatbot  
completion = client.chat.completions.create(
    model="gpt-5.4-mini",
    messages=[
        {
            "role": "system", # Can be system, user, assistant, tool
            "content": "You are Marv, a chatbot that reluctantly answers questions with sarcastic responses."
        },
        {
            "role": "user",
            "content": "Could you explain briefly what a black hole is?"
        }
    ],
    max_completion_tokens=250,
    temperature=1,
    seed=42
)

print(completion.choices[0].message.content)
````

````python
# Creating a sarcastic chatbot (streaming) 
stream = client.chat.completions.create(
    model="gpt-5.4-mini",
    messages=[
        {
            "role": "system", # Can be system, user, assistant, tool
            "content": "You are Marv, a chatbot that reluctantly answers questions with sarcastic responses."
        },
        {
            "role": "user",
            "content": "Could you explain briefly what a black hole is?"
        }
    ],
    max_completion_tokens=250,
    temperature=0.5,
    seed=42,
    stream=True
)

for chunk in stream:
    delta = chunk.choices[0].delta.content
    if delta:
        print(delta, end="")

print()
````

## LangChain Expression Language (LCEL)

````python
from dotenv import load_dotenv

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import CommaSeparatedListOutputParser
from langchain_openai import ChatOpenAI

load_dotenv()

chat = ChatOpenAI(
    model="gpt-5.4-mini", 
    temperature=0.5,
    max_completion_tokens=250,
    model_kwargs={"seed": 42}
)
````

````python
list_instructions = CommaSeparatedListOutputParser().get_format_instructions()
list_instructions
````

````python
chat_template = ChatPromptTemplate.from_messages(
    [
        ("human", "I've recently adopted a {pet_type}. Could you suggest three {pet_type} names?\n" + list_instructions),
    ]
)

chat_template
````

````python
print(chat_template.messages[0].prompt.template)
````

````python
list_output_parser = CommaSeparatedListOutputParser()
````

We can execute LangChain Expression Language as a separated set of results:

* Chat template result -> Chat result -> Output parser

````python
# LangChain Expression Language
chat_template_result = chat_template.invoke({"pet_type": "dog"})
chat_template_result
````

````python
chat_result = chat.invoke(chat_template_result)
chat_result
````

````python
list_output_parser.invoke(chat_result)
````

Using LCEL, we can obtain the same result in one line using the Pipe (`|`) symbol:

* The Pipe symbol links elements in the expression language
* So, the output of an element will be the input for the next component

````python
chain = chat_template | chat | list_output_parser
````

````python
chain.invoke({"pet_type": "dog"})
````

````python
from dotenv import load_dotenv

from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

load_dotenv()

chat = ChatOpenAI(
    model="gpt-5.4-mini", 
    temperature=0.5,
    max_completion_tokens=250,
    model_kwargs={"seed": 42}
)
````

````python
chain = chat_template | chat
````

````python
# Invoke calls
chain.invoke({"pet_type": "dog", "pet_breed": "shepherd"})
````

````python
%%time
# Batch calls (parallel requests)
batch = [
    {"pet_type": "dog", "pet_breed": "shepherd"},
    {"pet_type": "dragon", "pet_breed": "night fury"},
]
chain.batch(batch)
````

## Streaming Calls

The `.stream()` method produces a `generator` function:

* These `generator` functions allow for defining functions that behave like iterators, allowing us to loop over their output

* This is achieved through a `yield` statement that produces a sequence of values, instead of a `return` statement for standard functions

* `generator` functions have a single-use constraint to enhance efficiency

* Ideal for streaming responses from an LLM

Using the same code as above (batch calls).

````python
chain.invoke({"pet_type": "dog", "pet_breed": "shepherd"})
````

````python
chain.stream({"pet_type": "dog", "pet_breed": "shepherd"})
````

````python
response = chain.stream({"pet_type": "dog", "pet_breed": "shepherd"})
````

````python
# Execute the generator function several times
next(response)
````

````python
response = chain.stream({"pet_type": "dog", "pet_breed": "shepherd"})

for chunk in response:
    print(chunk.content, end="")
````

````python
response = chain.stream({"pet_type": "dog", "pet_breed": "husky"})

for chunk in response:
    print(chunk.content, end="")
````

## Piping Chains and the `RunnablePassthrough` 

````python
from dotenv import load_dotenv

from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_openai import ChatOpenAI

load_dotenv()
````

````python
RunnablePassthrough().invoke("Hi")
````

````python
RunnablePassthrough().invoke([1, 2, 3]) # Similar to an identity function
````

**Task**:

* Let's create a chain that lists the most essential tools for a given profession

* We'll feed this output to a second chain that suggests effective strategies for mastering these tools

````python
chat_template_tools = ChatPromptTemplate.from_template(
    """
    What are the five most important tools a {job_title} needs?

    Answer only by listing the tools, without any other text.
    """
)

chat_template_strategy = ChatPromptTemplate.from_template(
    """
    Considering the tools provided, develop a strategy for effectively learning and mastering them:
    
    {tools}
    """
)
````

````python
chat_template_tools
````

````python
chat_template_strategy
````

````python
chat = ChatOpenAI(
    model="gpt-5.4-mini", 
    temperature=0.5,
    max_completion_tokens=2000,
    model_kwargs={
        "seed": 42,
        "reasoning_effort": "low",
    }
)
````

````python
string_parser = StrOutputParser()
````

Create the pipeline of chains:

````python
chain_tools = chat_template_tools | chat | string_parser
````

````python
chain_strategy = chat_template_strategy | chat | string_parser
````

````python
print(chain_tools.invoke({"job_title": "data engineer"}))
````

````python
chain_strategy.invoke({"tools": """
SQL
Python
Apache Spark
Airflow
Git
"""})
````

Merging both chains:

* The first component of `chain_strategy` accepts a dictionary of `{"tools": "str output from chain_tools"}`

* So, this dictionary must be created at the end of `chain_tools`

* We use the `RunnablePassthrough()` class 

* The flow:

    - We pass the `{"job_title": "data engineer"}` to `chat_template_tools`

    - This creates a `ChatPromptValue` object which is input into the `chat` 

    - The `ChatOpenAI` class generates an `AIMessage`

    - The `AIMessage` is an argument used in the `string_parser` to provide a string output

    - The resulting string is used as input for the `RunnablePassthrough` class

    - By mirroring its input, it produces the dictionary `{"tools": "string output"}`

````python
new_chain_tools = chat_template_tools | chat | string_parser | {"tools": RunnablePassthrough()}
````

````python
new_chain_tools.invoke({"job_title": "data engineer"})
````

````python
combined_chain = new_chain_tools | chain_strategy
print(combined_chain.invoke({"job_title": "data engineer"}))
````

````python
chain_long = (
    chat_template_tools
    | chat
    | string_parser
    | {"tools": RunnablePassthrough()}
    | chat_template_strategy
    | chat
    | string_parser
)
print(chain_long.invoke({"job_title": "machine learning engineer"}))
````

## `RunnableParallel`

**Task**:

* A chain suggesting three intermediate-level programming language books

* A chain suggesting three projects on the same programming language at the same level

* These chains will be running simultaneously


````python
from dotenv import load_dotenv

from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableParallel
from langchain_openai import ChatOpenAI

load_dotenv()
````

````python
chat_template_books = ChatPromptTemplate.from_template(
    """
    Suggest three of the best intermediate-level {programming_language} books.

    Answer only by listing the books, without any other text.
    """
)

chat_template_projects = ChatPromptTemplate.from_template(
    """
    Suggest three interesting {programming_language} projects suitable for 
    intermediate-level programmers.

    Answer only by listing the projects, without any other text.
    """
)
````

````python
chat = ChatOpenAI(
    model="gpt-5.4-mini", 
    temperature=0.5,
    max_completion_tokens=2000,
    model_kwargs={"seed": 42}
)
````

````python
string_parser = StrOutputParser()
````

````python
# Create the chains
chain_books = chat_template_books | chat | string_parser
chain_projects = chat_template_projects | chat | string_parser
````

````python
# Parallel execution
chain_parallel = RunnableParallel({"books": chain_books, "projects": chain_projects})
````

````python
chain_parallel.invoke({"programming_language": "python"})
````

````python
chain_parallel.get_graph().print_ascii()
````

**Batch vs RunnableParallel**:

* `batch()`: Allows us to invoke the same `Runnable` with different input values

* `RunnableParallel`: Allows us to execute several `Runnable` using identical input values

````python
%%time
chain_books.invoke({"programming_language": "python"})
````

````python
%%time
chain_projects.invoke({"programming_language": "python"})
````

````python
%%time
chain_parallel.invoke({"programming_language": "python"})
````

## `RunnableLambda`

````python
find_sum([1, 2, 5])
````

````python
find_square = lambda x: x**2
````

Wrapping them in the `RunnableLambda` function

````python
from langchain_core.runnables import RunnableLambda
````

````python
runnable_sum = RunnableLambda(lambda x: sum(x))
````

````python
runnable_sum.invoke([1, 2, 5])
````

## The `@chain` Decorator

The `@chain` decorator wraps any function in a `RunnableLambda` function

* It's an elegant and intuitive way

* The function's name is also set as the name of the runnable

````python
from langchain_core.runnables import chain, RunnableLambda
````

````python
def find_sum(x):
    return sum(x)
````

````python
def find_square(x):
    return x**2
````

````python
# Convert to Runnable and create a chain 
chain = RunnableLambda(find_sum) | RunnableLambda(find_square)
````

````python
chain.invoke([1, 2, 5])
````

````python
# Convert to Runnable and create a chain using the @chain decorator

@chain
def runnable_sum(x):
    return sum(x)

@chain
def runnable_square(x):
    return x**2
````

````python
print(type(runnable_sum))
print(type(runnable_square))
````

````python
chain_decorator = runnable_sum | runnable_square
````

````python
chain_decorator.invoke([1, 2, 5])
````
